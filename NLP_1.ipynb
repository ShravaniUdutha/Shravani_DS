```python
def re_exp(text):
    print('original text:',text)
    words=text.split()
    print(words)
    re_punct=re.compile('[%s]'%re.escape(string.punctuation))
    w=[re_punct.sub('',w) for w in words]
    sent=''.join(w)
    print(w)
    print('preprocessed text:',sent)

re_exp("hello Chitty******")
```

**Normalisation**

- convert to lower case

```python
def text_cleaning(text1):
    words=text1.split() #divide into words
    re_punct=re.compile('[%s]'%re.compile(string.punctuation))
    w=[re_punct.sub('',w) for w in words] #remoiving punctuations
    w=[i.lower() for i in w] #lowering the words
    return(w)
```
